```python
# contineuas float value
con_f_val = []
for i, j in df.dtypes.items():
    if j == float:
        con_f_val.append(i)
con_f_val
```

```python
for i in con_f_val:
    print(i, " => ", df[i].isna().sum())
```

```python
for i in con_f_val:
    df[i].fillna(df[i].mean(), inplace=True)
```

```python
for i in con_f_val:
    print(i, " => ", df[i].isna().sum())
```

```python
con_i_val = df.columns
```

```python
con_i_val
```

```python
con_i_val = []
for i, j in df.dtypes.items():
    if j != float:
        con_i_val.append(i)
con_i_val
```

```python
for i in con_i_val:
    df[i].fillna(df[i].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
df.isna().sum().plot(kind='bar', figsize=(20, 8))
```

```python
df.head()
```

```python
def defending(data):
    return int(round((data[['Marking', "StandingTackle", "SlidingTackle", "Interceptions"]].mean()).mean()))

def mental_strong(data):
    return int(round((data[['Aggression', "Vision"]].mean()).mean()))

def power(data):
    return int(round((data[['Stamina', "Strength", "ShotPower", "Jumping", "LongShots"]].mean()).mean()))

def strike(data):
    return int(round((data[['Dribbling', "Finishing", "HeadingAccuracy", "FKAccuracy", "Volleys", "Penalties", "Composure"]].mean()).mean()))

def keeper(data):
    return int(round((data[['GKDiving', "GKHandling", "GKKicking", "GKPositioning", "GKReflexes"]].mean()).mean()))

def mobility(data):
    return int(round((data[['Acceleration', "SprintSpeed", "Agility", "Reactions", "Balance"]].mean()).mean()))

def passing(data):
    return int(round((data[['ShortPassing', "LongPassing", "Crossing"]].mean()).mean()))
```

```python
df.rename(columns={"Club Logo":"Club_Logo"}, inplace=True)
```

```python
df['defending'] = df.apply(defending, axis=1)
```

```python
df.head()
```

```python
df['mental_strong'] = df.apply(mental_strong, axis=1)
df['power'] = df.apply(power, axis=1)
df['strike'] = df.apply(strike, axis=1)
df['keeper'] = df.apply(keeper, axis=1)
df['mobility'] = df.apply(mobility, axis=1)
df['passing'] = df.apply(passing, axis=1)
```
