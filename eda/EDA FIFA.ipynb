## Feature Engineering

```python
df['Club'].fillna('No Club', inplace=True)
```

```python
df['Club'].isna().sum()
```

```python
df['Position'].isna().sum()
```

```python
df['Position'].value_counts()
```

```python
df['Position'].fillna('ST', inplace=True)
```

```python
df['Position'].isna().sum()
```

```python
df.dtypes
```

```python
# contineuas float value
con_f_val = []
for i, j in df.dtypes.items():
    if j == float:
        con_f_val.append(i)
con_f_val
```

```python
for i in con_f_val:
    print(i, " => ", df[i].isna().sum())
```

```python
for i in con_f_val:
    df[i].fillna(df[i].mean(), inplace=True)
```

```python
for i in con_f_val:
    print(i, " => ", df[i].isna().sum())
```

```python
con_i_val = df.columns
```

```python
con_i_val
```

```python
con_i_val = []
for i, j in df.dtypes.items():
    if j != float:
        con_i_val.append(i)
con_i_val
```

```python
for i in con_i_val:
    df[i].fillna(df[i].mode()[0], inplace=True)
```
